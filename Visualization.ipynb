```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('teen_phone_addiction_dataset.csv')
```

```python
numeric_cols = df.select_dtypes(include = np.number)

for col in numeric_cols:
    sns.kdeplot(data = df , x = col , fill = True)
    plt.show()
```

```python
addiction_counts = df['Addiction_Level'].value_counts(normalize=True)
plt.figure(figsize=(18, 6))
sns.barplot(x=addiction_counts.index, y=addiction_counts.values)
plt.title('Normalized Distribution of Addiction Levels')
plt.xlabel('Addiction Level')
plt.ylabel('Proportion')
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
df['Addiction_Level_yeojohnson'] = pt.fit_transform(df[['Addiction_Level']])


sns.kdeplot(data = df , x = 'Addiction_Level_yeojohnson' , fill = True)
plt.show()
```
